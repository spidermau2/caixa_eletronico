```python
#Objetivo: Programa em python que simula operações básicas de um caixa eletrônico
        
#---------------------------------------------------------------------------------------

from IPython.display import clear_output

saldo = 1000.00

opcao = 0

saida = 0

print("Bem vindo ao Caixa Eletrônico!\n")
input("Pressione enter para continuar...")

while saida == 0:
    clear_output()
    
    opcao = int(input("O que deseja fazer?\n1- Verificar saldo\n2- Depositar dinheiro\n3- Sacar dinheiro\n4- Sair\nEscolha uma opção (1-4)"))
    
    clear_output()
    
    if opcao == 1:
#     Verificar saldo:
#     - Ao escolher essa opção, deve exibir saldo atual.
        print(f"Seu saldo é: R${saldo:.2f}")
        input("Pressione enter para continuar...")
        
    elif opcao == 2:
#     Depositar saldo:
#     - Usuário deve ser capaz de inserir quantia na conta
#     - A quantia deve ser positiva
#     - Após um depósito bem sucedido, o saldo deve ser atualizado e uma msg de confirmação deve ser exibida     

        valor_deposito = 0   
        questao = 1
    
        while questao != 0:       
            valor_deposito = float(input(f"Insira valor para deposito (Valor atual: R${saldo:.2f}): "))
            while valor_deposito < 1:
                clear_output()
                valor_deposito = float(input("0 ou números negativos não permitido, reinsira: "))
                clear_output()
            questao = float(input(f"Você inseriu R${valor_deposito:.2f}, seu novo total é: R${saldo + valor_deposito:.2f}. Deseja continuar? (0-Sim/1-Não)"))

        if questao == 0:
            saldo = saldo + valor_deposito              
        continue

    elif opcao == 3:
#     Sacar dinheiro:
#     - O usuário deve ser capaz de inserir quantia para sacar da conta
#     - A quantia deve ser positiva e não deve exceder saldo atualizado
#     - Após um saque bem sucedido, o saldo deve ser atualizado e uma msg de confirmação deve ser exibida
        valor_saque = 0   
        questao = 1

        while questao != 0:       
            clear_output()
            valor_saque = float(input(f"Insira valor para saque (Valor atual: R${saldo:.2f}): "))
            while valor_saque < 0 or valor_saque > saldo:
                clear_output()
                valor_saque = float(input("0, números negativos ou retirada maior que valor em conta não permitido, reinsira: "))
                clear_output()
            questao = float(input(f"Você retirou R${valor_saque:.2f}, seu novo total é: R${saldo - valor_saque:.2f}. Deseja continuar? (0-Sim/1-Não)"))
            if questao == 0:
                saldo = saldo - valor_saque
        continue
    elif opcao == 4:
        saida = 1
        print("\nFim do programa.\n")
```

```python
#Versão com funções e try except

from IPython.display import clear_output

def exibir_saldo(saldo):
    print(f"Seu saldo é: R${saldo:.2f}")
    input("Pressione enter para continuar...")

def depositar(saldo):
    valor_deposito = 0   
    questao = 1
    
    while questao != 0:
        try:
            valor_deposito = float(input(f"Insira valor para depósito (Valor atual: R${saldo:.2f}): "))
        except ValueError:
            print("Valor inválido. Insira um número válido.")
            continue
        
        while valor_deposito < 1:
            valor_deposito = float(input("Valor inválido, reinsira: "))
            
        questao = float(input(f"Você depositou R${valor_deposito:.2f}, seu novo total é: R${saldo + valor_deposito:.2f}. Deseja continuar? (0-Sim/1-Não)"))

    if questao == 0:
        saldo = saldo + valor_deposito
    return saldo

def sacar(saldo):
    valor_saque = 0   
    questao = 1

    while questao != 0:
        try:
            valor_saque = float(input(f"Insira valor para saque (Valor atual: R${saldo:.2f}): "))
        except ValueError:
            print("Valor inválido. Insira um número válido.")
            continue
        
        while valor_saque < 0 or valor_saque > saldo:
            valor_saque = float(input("Valor inválido, reinsira: "))
            
        questao = float(input(f"Você retirou R${valor_saque:.2f}, seu novo total é: R${saldo - valor_saque:.2f}. Deseja continuar? (0-Sim/1-Não)"))
        
        if questao == 0:
            saldo = saldo - valor_saque

    return saldo

saldo = 1000.00
opcao = 0
saida = 0

print("Bem-vindo ao Caixa Eletrônico!\n")
input("Pressione enter para continuar...")

while saida == 0:
    clear_output()
    opcao = int(input("O que deseja fazer?\n1- Verificar saldo\n2- Depositar dinheiro\n3- Sacar dinheiro\n4- Sair\nEscolha uma opção (1-4)"))
    
    if opcao == 1:
        clear_output()
        exibir_saldo(saldo)
        
    elif opcao == 2:
        clear_output()
        saldo = depositar(saldo)
        
    elif opcao == 3:
        clear_output()
        saldo = sacar(saldo)
        
    elif opcao == 4:
        saida = 1
        print("\nFim do programa.\n")
```
